```python
import pandas as pd
```

```python
#Load Data
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
```

```python
# Group your data
air_data_grouped = air_data.groupby('sensor.name')
```

```python
# Summary stats on PM2.5

# Groups the PM2.5 data into different stastical stats

air_data_pm25 = air_data_grouped.agg(max = ('pm2.5_atm', 'max'),
                                     min = ('pm2.5_atm', 'min'),
                                     median = ('pm2.5_atm', 'median'),
                                     mean = ('pm2.5_atm', 'max'))
# Find the top 5 Mean

top5_mean_pm25 = (
    air_data_pm25.sort_values(by = 'mean', ascending = False)
)

top5_mean_pm25.head(5)
```

```python
# Find the top 5 Median

top5_median_pm25 = (
    air_data_pm25.sort_values(by = 'median', ascending = False)
)

top5_median_pm25.head(5)
```

```python
# Summary stats on PM 10.0

# Groups the PM10.0 data into different stastical stats

air_data_pm10 = air_data_grouped.agg(max = ('pm10.0_atm', 'max'),
                                     min = ('pm10.0_atm', 'min'),
                                     median = ('pm10.0_atm', 'median'),
                                     mean = ('pm10.0_atm', 'max'))
# Find the top 5 Mean

top5_mean_pm10 = (
    air_data_pm10.sort_values(by = 'mean', ascending = False)
)

top5_mean_pm10.head(5)
```

```python
# Find the top 5 Median

top5_median_pm10 = (
    air_data_pm10.sort_values(by = 'median', ascending = False)
)

top5_median_pm10.head(5)
```

```python
# Summary stats on VOC

# Groups the VOC data into different stastical stats

air_data_voc = air_data_grouped.agg(max = ('voc', 'max'),
                                     min = ('voc', 'min'),
                                     median = ('voc', 'median'),
                                     mean = ('voc', 'max'))
# Find top 5

top5_mean_voc = (
    air_data_voc.sort_values(by = 'mean', ascending = False)
)

top5_mean_voc.head(5)
```

```python
#Assign temperature to different sections

def assign_temperature(temperature): 
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 50 < temperature <= 70:
        return 'Warm'
    elif 70 < temperature <= 1000:
        return 'Hot'
    else:
        return 'No Data' 


air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

#Assign humidity to different sections

def assign_humidity(humidity): 
    if humidity < 50:
        return 'Low Humidity'
    elif 50 <= humidity <= 80:
        return 'High Humidity'
    elif 80 <= humidity <= 10000:
        return 'Very High Humidity'
    else:
        return 'No Data'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

#Assign Altitude to different sections

def assign_altitude(altitude): 
    if altitude < 1000:
        return 'Low'
    elif 1000 <= altitude <= 2500:
        return 'Medium'
    else: 
        return 'High'

air_data['altitude_level'] = air_data['sensor.altitude'].apply(assign_altitude)
```
